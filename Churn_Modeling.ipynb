```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Churn_Modelling.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

### There is no missing values in dataset so lets create dummy variables for the categorical data.

```python
# There is no use of rownumber column so remove the column.
dataset.drop(columns=['RowNumber'],inplace=True)
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.head(10)
```

# Handling the Outliers

```python
from scipy import stats
```

```python
pd.DataFrame(stats.zscore(dataset))
```

```python
zscoreData = np.abs(stats.zscore(dataset))
```

```python
dataset[(zscoreData > 3).all(axis = 1)]
```

```python
dataset[(zscoreData < 0.3).all(axis = 1)]
```

# Spliting the feature data and the target data

```python
X=dataset.drop("Exited",axis=1)
y=dataset["Exited"]
```

```python
X.head()
```

```python
y.head()
```

# Split the Training Data and Test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
y_test.shape
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc.fit(X_train)
```

```python
# changing the Training data
X_train_scaled = sc.transform(X_train)
```

```python
pd.DataFrame(X_train_scaled).head()
```

```python
# Change the Testing data as well
X_test_scaled = sc.transform(X_test)
```

# Model Building

```python
from sklearn.svm import SVC
```

```python
svmClassifier = SVC(C= 1.0, kernel='linear', random_state= 0)
```

## Train the model

```python
svmClassifier.fit(X_train_scaled,y_train)
```

## Test the result

```python
y_pred=svmClassifier.predict(X_test_scaled)
```

```python
y_pred
```

##  Accuracy_score

```python
from sklearn.metrics import  accuracy_score
```

```python
print("Accuracy Score: ", accuracy_score(y_test,y_pred)*100)
```

# Accuracy Score of support vector classifier is  78.8 %
